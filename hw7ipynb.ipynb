```python
from Bio import SeqIO
record = SeqIO.read("/content/sequence.fasta", "fasta")
s_gene = record.seq[21562:25384]
with open("/content/seq1.fasta", "w") as f:
    f.write(f">S_gene_{record.id}\n{s_gene}\n")
```

```python
files = ["seq1.fasta", "/content/sequence (1).fasta", "seq3.fasta","seq4.fasta"]
output_file = "all_seq3.fasta"

with open(output_file, "w") as out_handle:
    for file in files:
        for record in SeqIO.parse(file, "fasta"):
            SeqIO.write(record, out_handle, "fasta")
```
