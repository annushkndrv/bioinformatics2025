# Филогенетическое дерево

Я выбрала гексокиназу 1. Она катализирует фосфорилирование различных гексоз, таких как D-глюкоза, D-глюкозамин, 
D-фруктоза, D-манноза и 2-дезокси-D-глюкоза, до гексозо-6-фосфата (D-глюкозо-6-фосфат, D-глюкозамин-6-фосфат, D-фруктозо-6-фосфат, D-маннозо-6-фосфат и 2-дезокси-D-глюкозо-6-фосфат соответственно). 
Участвует в начальной стадии гликолиза, катализируя фосфорилирование D-глюкозы до D-глюкозо-6-фосфата).
Участвует во врожденном иммунитете и воспалении, действуя как рецептор распознавания паттернов бактериального пептидогликана. 
При попадании в цитозоль N-ацетил-D-глюкозамин, входящий в состав бактериального пептидогликана, 
ингибирует активность гексокиназы HK1 и вызывает ее диссоциацию с внешней мембраны митохондрий, 
тем самым активируя инфламмасому NLRP3.
